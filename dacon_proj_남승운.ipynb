```python
df['인수여부']=df['인수여부'].map({'Yes': 1, 'No': 0}) # Yes는 1, No는 0
df.head()
```

```python
df['상장여부']=df['상장여부'].map({'Yes': 1, 'No': 0}) # Yes는 1, No는 0
df.head()
```

```python
df['분야']=df['분야'].fillna('기타') # 결측값은 기타로 채운다
df.info()
```

```python
from sklearn.linear_model import LinearRegression

train = df[df['직원 수'].notna()] # 결측값 없는 행 -> 학습
test = df[df['직원 수'].isna()] # 결측값 행 -> 예측

features = ['총 투자금(억원)', '연매출(억원)'] # 활용할 변수
X_train = train[features]
y_train = train['직원 수']
X_test = test[features]

model = LinearRegression() # 선형 회귀
model.fit(X_train, y_train)
preds = model.predict(X_test)

df.loc[df['직원 수'].isna(), '직원 수'] = preds
df['직원 수'].describe()
```

```python
train = df[df['고객수(백만명)'].notna()] # 결측값 없는 행 -> 학습
test = df[df['고객수(백만명)'].isna()] # 결측값 행 -> 예측

features = ['분야_AI', '분야_게임', '분야_기술', '분야_기타', '분야_물류', '분야_에너지', 
            '분야_에듀테크', '분야_이커머스', '분야_푸드테크', '분야_핀테크', '분야_헬스케어', '연매출(억원)'] # 활용할 변수
X_train = train[features]
y_train = train['고객수(백만명)']
X_test = test[features]

model = LinearRegression() # 선형 회귀
model.fit(X_train, y_train)
preds = model.predict(X_test)

df.loc[df['고객수(백만명)'].isna(), '고객수(백만명)'] = preds
```

```python
# 평균값을 반환하는 함수, 결측값은 그대로 결측값, 6000 이상은 6000으로 반환
def mean_value(x):
    if pd.isna(x):
        return np.nan
    parts = x.split('-')
    if len(parts) == 2:
        start, end = x.split('-')
        return (int(start) + int(end)) / 2
    else:
        return 6000

df['기업가치(백억원)']=df['기업가치(백억원)'].apply(mean_value) # 함수 적용
```

```python
train = df[df['기업가치(백억원)'].notna()] # 결측값 없는 행 -> 학습
test = df[df['기업가치(백억원)'].isna()] # 결측값 행 -> 예측

features = ['총 투자금(억원)', '연매출(억원)', '직원 수'] # 활용할 변수
X_train = train[features]
y_train = train['기업가치(백억원)']
X_test = test[features]

model = LinearRegression()
model.fit(X_train, y_train)
preds = model.predict(X_test)

df.loc[df['기업가치(백억원)'].isna(), '기업가치(백억원)'] = preds
```
